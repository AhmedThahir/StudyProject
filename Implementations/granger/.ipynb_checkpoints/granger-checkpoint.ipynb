# Hypothesis

Market Value is a function of performance and age.

$$
V \propto P \\
V \frac{1}{\propto} A
$$

# Idea

Maybe, we can apply this in the reverse way to see if a player is worth to buy, based on performance and age.

# Reasoning

### Granger Causality

It helps understand a pseudo-causality(not necessarily true causality) using time-series data.

$$
y_t = y_{t-1} + x_{t-1}
$$

If $p$ value < 0.05, then it is significant.

### Base Value

The base/'zero' value of every player must be their initial value , as every player starts off from a different initial value. This is not just restricted to talent/skill, due to factors such as nationality. Moreover, it helps accomodate the market value of a player who started off with a high value won't really go < 1M.

---

# Quick Case Study

For eg, Jack Wilshere. His market value is currently (as of May 17 2021) is €2.00M, despite injuries and poor performances over the past few years. Does this mean that he is better than a player who is currently only €1.00M?

**Not** necessarily. This is because his value has gone down from the initial year, indicating the worsening of his performances. The other player might have started at 100K, and now is at 1M, which means that **might** be they are better than Jack Wilshere, which is reflected by their increase of value by 900k. We can conclude only from further analyses, rather than just current absolute market value.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg'] # makes everything svg by default
%matplotlib inline

from datetime import datetime
from datetime import date

from statsmodels.tsa.stattools import grangercausalitytests
```

# Market Value

```python
# create an Empty DataFrame object
mv = pd.DataFrame(
    columns = ['Date', 'Team', 'Market_Value'],
    data = [
        ['04-10-2004', 'Manchester United', 18.00],
        ['11-02-2005', 'Manchester United', 22.80],
        ['18-10-2005', 'Manchester United', 26.50],
        ['14-06-2006', 'Manchester United', 25.00],
        ['25-04-2007', 'Manchester United', 34.00],
        ['04-02-2008', 'Manchester United', 50.00],
        ['08-04-2008', 'Manchester United', 55.00],
        ['04-07-2008', 'Manchester United', 60.00],
        ['22-07-2009', 'Real Madrid', 70.00],
        ['03-12-2009', 'Real Madrid', 70.00],
        ['07-01-2010', 'Real Madrid', 75.00],
        ['12-04-2010', 'Real Madrid', 75.00],
        ['27-08-2010', 'Real Madrid', 90.00],
        ['04-02-2011', 'Real Madrid', 90.00],
        ['29-07-2011', 'Real Madrid', 90.00],
        ['04-03-2012', 'Real Madrid', 90.00],
        ['07-08-2012', 'Real Madrid', 100.00],
        ['10-01-2013', 'Real Madrid', 100.00],
        ['12-06-2013', 'Real Madrid', 100.00],
        ['23-01-2014', 'Real Madrid', 100.00],
        ['20-07-2014', 'Real Madrid', 100.00],
        ['22-10-2014', 'Real Madrid', 100.00],
        ['22-10-2014', 'Real Madrid', 120.00],
        ['23-01-2015', 'Real Madrid', 120.00],
        ['01-07-2015', 'Real Madrid', 120.00],
        ['22-11-2015', 'Real Madrid', 110.00],
        ['22-02-2016', 'Real Madrid', 110.00],
        ['01-06-2016', 'Real Madrid', 110.00],
        ['15-07-2016', 'Real Madrid', 110.00],
        ['24-01-2017', 'Real Madrid', 100.00],
        ['05-06-2017', 'Real Madrid', 100.00],
        ['01-01-2018', 'Real Madrid', 120.00],
        ['30-05-2018', 'Real Madrid', 100.00],
        ['20-12-2018', 'Juventus FC', 100.00],
        ['06-06-2019', 'Juventus FC', 90.00],
        ['06-06-2019', 'Juventus FC', 90.00],
        ['12-12-2019', 'Juventus FC', 75.00],
        ['08-04-2020', 'Juventus FC', 60.00],
        ['25-08-2020', 'Juventus FC', 60.00],
        ['29-12-2020', 'Juventus FC', 60.00],
        ['23-03-2021', 'Juventus FC', 50.00],
        ['01-06-2021', 'Juventus FC', 45.00],
        ['23-12-2021', 'Manchester United', 35.00]
    ]
)

mv['Date']= pd.to_datetime(mv['Date'], format='%d-%m-%Y')
# mv["Year"] = mv["Date"].dt.year

mv["Season"] = mv["Date"].dt.year

mv["Age"] = ''
display(mv)
```

```python
def calc_Season(date):
    if(date.month > 7):
        start_of_season = date.year
    else:
        start_of_season = date.year - 1
        
    end_of_season = start_of_season + 1
    season = str(start_of_season) + "/" + str(end_of_season)
    return season

def calculateAge(current, dob):
    age = current.year - dob.year - ((current.month, current.day) <
         (dob.month, dob.day))
 
    return age
```

```python
dob = '05/02/1985'
dob = datetime.strptime(dob, '%d/%m/%Y')

for i in range(mv.shape[0]):
    mv['Age'].iloc[i] = calculateAge(mv['Date'].iloc[i], dob)
    mv['Season'].iloc[i] = calc_Season(mv['Date'].iloc[i])

display(mv)
```

```python
mv["Value_Diff_From_First"] = mv["Market_Value"] - mv["Market_Value"].iloc[0]

# mv = mv.iloc[1: , : ] # delete the first record, as it won't have deviation
# commented cuz I will be averaging for the season (can't think of how to explain)

mv['Age'] = mv['Age'].astype(int) # otherwise causing an error

display(mv)
```

```python
meanCols = {
#     "Market_Value": "Mean_Market_Value",
    "Value_Diff_From_First": "Mean_Value_Diff_From_First",
    "Age": "Mean_Age"
}

mv_mean = mv[["Season", "Age", "Value_Diff_From_First"]].groupby(
    ["Season"],
    as_index = False
).mean().rename(columns=meanCols)

# rounding to 2 places
mv_mean["Mean_Value_Diff_From_First"] = np.around(mv_mean["Mean_Value_Diff_From_First"], 2)
mv_mean["Mean_Age"] = np.around(mv_mean["Mean_Age"], 2)

display(mv_mean)
```

# Performance

```python
perf = pd.DataFrame(
    columns = ['Season', 'Appearances', 'Goals', 'Assists'],
    data = [
        ["2021/2022", 39, 24, 3],
        ["2020/2021", 44, 36, 4],
        ["2019/2020", 46, 37, 7],
        ["2018/2019", 43, 28, 11],
        ["2017/2018", 44, 44, 8],
        ["2016/2017", 46, 42, 12],
        ["2015/2016", 48, 51, 15],
        ["2014/2015", 54, 61, 23],
        ["2013/2014", 47, 51, 17],
        ["2012/2013", 55, 55, 13],
        ["2011/2012", 55, 60, 15],
        ["2010/2011", 54, 53, 18],
        ["2009/2010", 35, 33, 10],
        ["2008/2009", 53, 26, 12],
        ["2007/2008", 49, 42, 8],
        ["2006/2007", 53, 23, 22],
        ["2005/2006", 47, 12, 9],
        ["2004/2005", 50, 9, 9],
        ["2003/2004", 40, 6, 8],
        ["2002/2003", 31, 5, 6],
        ["2001/2002", 31, 5, 6]
    ]
)

perf = perf.sort_values("Season").reset_index(drop=True)

perf["Goal_Contributions"] = perf["Goals"] + perf["Assists"]
display(perf)
```

# Merged

```python
merged = pd.merge(
    mv_mean,
    perf
)

display(merged)
```

```python
merged["Input"] = merged["Goal_Contributions"] - merged["Mean_Age"]

display(merged)
```

# Analysis

```python
merged.plot(x="Season", ylabel="Feature", title="Features over Time")
```

```python
def granger_analysis(y, x, lag=1):
    print(y, "as a function of", x)
    print()
    
    # df = pd.DataFrame(columns=['y', 'x'])
    plot_df = merged[["Season", x, y]]

    # don't sort cuz it will affect granger causality
    plot_df.plot(x="Season", ylabel="Feature", title=y+" vs "+x)
    plt.show()
    
    granger_df = merged[[y, x]] # granger library requires y,x format
    grangercausalitytests(granger_df, lag)
```

```python
granger_analysis("Mean_Value_Diff_From_First", "Goal_Contributions")
```

```python
granger_analysis("Goal_Contributions", "Mean_Value_Diff_From_First")
```

```python
granger_analysis("Mean_Value_Diff_From_First", "Input")
```
