# Overall Rating vs Value

Using only FIFA 22 Data
(otherwise, we can't use simple OLS)

# Limitations

- CPI is not effective here
- This is only for a single year
- This only works for players with value $>= €10M$

# Conclusions
- Linear relationship between rating and value for big players
- There are many good under-valued players.

```python
import warnings
warnings.filterwarnings('ignore')
from IPython.display import display, Math, Latex

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scale = StandardScaler()

import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg'] # makes everything svg by default
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
columns = ["ID", "Name", "Value", "Overall"]
```

```python
fifa = list()

fifa = pd.read_csv("data/FIFA21.csv")[columns]
    
display(fifa.head())
```

```python
cpi = pd.read_csv("data/cpiEuro.csv")

cpi.rename(columns={ 
    cpi.columns[0]: "Year",
    cpi.columns[1]: "CPI"
}, inplace = True)

cpi["Year"] = cpi["Year"].str[:4]
cpi["Year"] = cpi["Year"].astype(int)

display(cpi.tail())
```

```python
ratings = pd.DataFrame(columns = columns)

query = "ID == 212198"

records = fifa
records.loc[:, "Year"] = 2021
ratings = ratings.append(records)

ratings["Value"] = ratings["Value"].str.replace('€', '')

m = ratings["Value"].str.endswith("M"), "Value"
k = ratings["Value"].str.endswith("K"), "Value"

ratings.loc[m] = ratings.loc[m].str.replace('M', '').astype(float)*1000
ratings.loc[k] = ratings.loc[k].str.replace('K', '').astype(float)


ratings["Year"] = ratings["Year"].astype(int)
ratings["Value"] = ratings["Value"].astype(float)
ratings["Overall"] = ratings["Overall"].astype(int)
ratings.reset_index(drop=True, inplace=True)

display(ratings)
```

```python
final = pd.merge(
    ratings, cpi
).sort_values("Value").reset_index(drop=True)

final = final.drop(index = 
                   final[
                       final["Value"] < 10000
                   ].index
                  ).reset_index(drop=True)
final["CorrectedValue"] = final["Value"]/final["CPI"]

# final[["Value", "CorrectedValue", "Overall"]]  = scale.fit_transform(final[["Value", "CorrectedValue", "Overall"]])
display(final)
```

```python
def analysis(xVar, yVar, deg = 1, fix = 1):
    title = yVar + " vs " + xVar

    x = final[[xVar]]
    y = final[[yVar]]

    poly = PolynomialFeatures(degree = deg)
    xPoly = poly.fit_transform(x)

    model = LinearRegression()
    model.fit(xPoly, y)
    prediction = model.predict(xPoly)
    
#     fixes
    np.around(prediction)
    if(fix == 1):
        np.where( prediction>100, prediction, 100 )

#     plots
    plt.plot(x, y, '.')
    plt.plot(x, prediction)

    # plot config
    plt.title(title)
    plt.xlabel(xVar), plt.ylabel(yVar)
    plt.legend([
        "Point", "Regression"
    ])

    plt.savefig(title + ".svg", dpi=300, bbox_inches = 'tight')

    rSquared = round(r2_score(prediction, y), 4)
    display(Latex(rf'$R^2$ Value $= {rSquared}$'))
    plt.show()
```

```python
analysis("Value", "Overall")
```

```python
analysis("CorrectedValue", "Overall")
```

```python
analysis("Value", "Overall", deg = 2)
```
