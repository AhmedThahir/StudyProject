# Overall Rating vs Value

Using only FIFA 22 Data
(otherwise, we can't use simple OLS)
Tried various of degrees of regression

# Hypothesis
$$
\text{Deviation of Player Rating from Mean} \propto \text{Relative Price}
$$

# Limitations

- CPI is not effective here
- Since there are only 5 years of data, we **cannot** do any form of
    - timeseries analysis
    - granger causality analysis
- This only works for players with value $\ge €10M$

# Conclusions
- Linear relationship between rating and value for big players
- There are many good under-valued players.

```python
import warnings
warnings.filterwarnings('ignore')
from IPython.display import display, Math, Latex

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scale = StandardScaler()

import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg'] # makes everything svg by default
%matplotlib inline

import pandas as pd
import numpy as np

rel = "../data/"
```

```python
columns = ["ID", "Name", "Value", "Overall"]

fifa = list()
for year in range(17, 22+1): # 17-22
    file = rel + "fifa/FIFA" + str(year) + ".csv"
    fifa.append(
        pd.read_csv(file)[columns]
    )
    
fifa[5].head()
```

```python
cpi = pd.read_csv(rel + "cpiEuro.csv")

cpi.rename(columns={ 
    cpi.columns[0]: "Year",
    cpi.columns[1]: "CPI"
}, inplace = True)

cpi["Year"] = cpi["Year"].str[:4]
cpi["Year"] = cpi["Year"].astype(int)

display(cpi.tail())
```

```python
ratings = pd.DataFrame(columns = columns)

query = "ID == 212198"

for i in range(0, 5+1):
    records = fifa[i]
    records.loc[:, "Year"] = 2017+i
    ratings = ratings.append(records)

ratings["Value"] = ratings["Value"].str.replace('€', '')

m = ratings["Value"].str.endswith("M"), "Value"
k = ratings["Value"].str.endswith("K"), "Value"

ratings.loc[m] = ratings.loc[m].str.replace('M', '').astype(float)*1000
ratings.loc[k] = ratings.loc[k].str.replace('K', '').astype(float)


ratings["Year"] = ratings["Year"].astype(int)
ratings["Value"] = ratings["Value"].astype(float)
ratings["Overall"] = ratings["Overall"].astype(int)
ratings.reset_index(drop=True, inplace=True)

display(ratings)
```

```python
merged = pd.merge(
    ratings[ratings["Value"] >= 10000],
    cpi
).sort_values("Value").reset_index(drop=True)

merged["CPIValue"] = merged["Value"]/merged["CPI"]

display(merged)
```

```python
def analysis(xVar, yVar, df = merged, degs = [1], alpha=0.30):
    title = yVar + " vs " + xVar

    df = df.sort_values(xVar)
    
    x = df[[xVar]]
    y = df[[yVar]]
    plt.plot(x, y, '.', alpha=alpha)
    legends = ["Point"]
    display(Latex
            (rf'$R^2$ Value for')
           )
    
    for deg in degs:
        legends.append(
            "Degree " + str(deg)
        )
        poly = PolynomialFeatures(degree = deg)
        xPoly = poly.fit_transform(x)

        model = LinearRegression()
        model.fit(xPoly, y)
        prediction = model.predict(xPoly)
    
    #     fixes
        np.around(prediction)
        np.where( prediction>100, prediction, 100 )

        plt.plot(x, prediction)

        # plot config
        rSquared = round(r2_score(prediction, y), 4)
        display(Latex(rf'Degree {deg} $= {rSquared}$'))
        
    plt.title(title)
    plt.xlabel(xVar), plt.ylabel(yVar)
    plt.legend(legends)
    plt.savefig(title + ".svg", dpi=300, bbox_inches = 'tight')
    plt.show()
```

```python
analysis("Value", "Overall", degs = [1, 2, 3, 4])
```

```python
boldLeft = '\033[1m'
boldRight = '\033[0m'

meanCols = {
    "Value": "MeanValue",
    "Overall": "MeanOverall",
    "CPI": "MeanCPI",
    "CPIValue": "MeanCPIValue"
}

mean = merged.groupby(["Year"], as_index=False).mean().rename(columns=meanCols)
display(mean)
```

```python
relative = pd.merge(
    merged, mean, on = "Year"
)#.sort_values("Value").reset_index(drop=True)

relative["RelativeMeanValue"] = relative["Value"]/relative["MeanValue"]
relative["DiffMeanOverall"] = relative["Overall"]-relative["MeanOverall"]

display(relative)
```

```python
analysis("RelativeMeanValue", "DiffMeanOverall", df=relative, degs=[1,2,3,4])
```
